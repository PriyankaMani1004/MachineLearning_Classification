```python
#importing the Libraies
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
```

```python
# Reading the Dataset
dataset = pd.read_csv('Social_Network_Ads.csv')
dataset
```

```python
dataset=pd.get_dummies(dataset,drop_first=True)
dataset=dataset.replace({True:1,False:0})
dataset
```

```python
dataset=dataset.drop("User ID",axis=1)
dataset
```

```python
dataset["Purchased"].value_counts()
```

```python
independent=dataset[['Age', 'EstimatedSalary','Gender_Male']]
independent
```

```python
dependent=dataset['Purchased']
dependent
```

```python
#split into training set and test
from sklearn.model_selection import train_test_split
X_train, X_test, Y_train, Y_test = train_test_split(independent, dependent, test_size = 1/3, random_state = 0)
X_train, X_test, Y_train, Y_test
```

```python
from sklearn.preprocessing import MinMaxScaler

scaler = MinMaxScaler()
X_train = scaler.fit_transform(X_train)
X_test = scaler.transform(X_test)
```

```python
#MultinominalNB
from sklearn.naive_bayes import MultinomialNB
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.0, 0.1, 0.5, 1.0, 10.0],
              'fit_prior': [True, False]}
grid = GridSearchCV(MultinomialNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
grid.fit(X_train, Y_train)

```

```python
y_pred = grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, y_pred)              
print(clf_report)
print(cm)
```

```python
#BernoulliNB
from sklearn.naive_bayes import BernoulliNB
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.0, 0.1, 0.5, 1.0, 10.0],
              'binarize': [0.0, 0.5, 1.0, None],
              'fit_prior': [True, False]}
grid = GridSearchCV(BernoulliNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
grid.fit(X_train, Y_train)
```

```python
y_pred = grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, y_pred)              
print(clf_report)
print(cm)
```

```python
#CategoricalNB
from sklearn.naive_bayes import CategoricalNB
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.0, 0.1, 0.5, 1.0, 10.0],              
              'fit_prior': [True, False]}
grid = GridSearchCV(CategoricalNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
grid.fit(X_train, Y_train)
```

```python
y_pred = grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, y_pred)              
print(clf_report)
print(cm)
```

```python
#ComplementNB
from sklearn.naive_bayes import ComplementNB
from sklearn.model_selection import GridSearchCV
param_grid = {'alpha': [0.0, 0.1, 0.5, 1.0, 10.0],
              'fit_prior': [True, False],
              'norm': [True, False]}
grid = GridSearchCV(ComplementNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
grid.fit(X_train, Y_train)
```

```python
y_pred = grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, y_pred)              
print(clf_report)
print(cm)
```

```python
#GaussianNB
from sklearn.naive_bayes import GaussianNB
from sklearn.model_selection import GridSearchCV
param_grid = {'var_smoothing': np.logspace(-11, -7, 10)}
grid = GridSearchCV(GaussianNB(), param_grid, refit = True, verbose = 3,n_jobs=-1,scoring='f1_weighted') 
   
grid.fit(X_train, Y_train)
```

```python
y_pred = grid.predict(X_test)
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(Y_test, y_pred)
from sklearn.metrics import classification_report
clf_report = classification_report(Y_test, y_pred)              
print(clf_report)
print(cm)
```
